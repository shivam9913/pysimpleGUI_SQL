```python
  def sakila():  
    
    sg.theme("NeutralBlue")
    title="Queries of Actor Table"
    
    layout=[[sg.Text("list the names of all english films", size=(30,2)), 
                sg.Button("Click", key="one",size=(10,2))], 
            [sg.Text("Display the first name and the last name of all actors in the actor table", size=(30,2)), 
                sg.Button("Click", key="two",size=(10,2))], 
            [sg.Text(" Display the first and the last names in a single column", size=(30,2)), 
                 sg.Button("Click", key="three",size=(10,2))], 
            [sg.Text("find all actors whose last names contains the word 'Gen'", size=(30,2)), 
                 sg.Button("Click", key="four",size=(10,2))], 
            [sg.Text(" Find the id, name of the actor with the first name 'Joe'", size=(30,2)), 
                sg.Button("Click", key="five",size=(10,2))], 
            [sg.Text(" List the last names of actors who is having highest frequency", size=(30,2)), 
                 sg.Button("Click", key="six",size=(10,2))]]
    
      
    
    window=sg.Window(title,layout)
    
    while True:
        event,value=window.read()
                
        if event == "one":
            query = "select film.title from film inner join language where film.language_id = language.language_id and language.name = 'English';"
            sakila_db(query)
        
        if event == "two":
            query = "select first_name, last_name from sakila.actor;"
            sakila_db(query)
        
        if event == "three":
            query = "select concat(first_name, ' ', last_name) as 'Actor Name' from sakila.actor;"
            sakila_db(query)
        
        if event == "four":
            query = "select * from sakila.actor where last_name like '%Gen%';"
            sakila_db(query)
        
        if event == "five":
            query = "select actor_id, first_name, last_name from sakila.actor where first_name = 'Joe';"
            sakila_db(query)
         
        if event == "six":
            query = "select last_name, count(*) Count from sakila.actor group by last_name having Count > 2 order by Count desc;"
            sakila_db(query)
        
       
                   
            window.close()
            
            
        if  event == sg.WIN_CLOSED:
            break
    window.close()  
```
